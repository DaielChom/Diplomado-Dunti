#### ONA:
Esta plataforma podría cumplir con los requisitos que requiere Geomatica, cumple con la recolección de los datos geográficos y el uso offline de los formularios, al parecer se puede hacer una instalación local como se indica en la sección [FAQ](https://help.ona.io/faq/general-FAQ/):
> Can I host my own version of Ona i.e. install a local deployment?
Yes, please see https://github.com/SEL-Columbia/formhub/wiki/Installation-and-Deployment. In some cases, we also work with clients to implement installs. Please email sales@ona.io with an overview of your needs.

En otra sección muestra:
>Sometimes data security laws mandate that data is hosted in country. For these instances, we can help setup and manage local instances of Ona for our clients.

La guía de instalación es netamente para Linux y esta desactualizada siendo Ubuntu 12 la versión más actualizada para la que se realizó la guía. al parecer ONA está desarrollada en Python y puede usar mongo, mysql, postgres como base de datos. lo extraño es que el repositorio hace referencia de formhub más que a ONA. Es de pago, tiene una versión gratis, la cual presenta un número limitado de proyectos privados, entre otras limitantes, las otras versiones son habilitadas con un pago mensual.

Parece ser que si existe un repositorio de [ona](https://github.com/onaio), se denomina a ella misma como derivada de formhub

#### Rotatorsurvey:
Es un software disponible únicamente para Windows y Android, permite la creacion de formularios, el uso offline, el análisis de los datos se realiza con gran diversidad de gráficas, permite exportar los datos obtenidos a Excel y Spss. Cuenta con una versión gratuita con limitación de encuestas, ofrece varios planes anuales de pago con distintas limitantes cada uno, dependiendo del tamaño del estudio de campo que se quiera realizar. no tiene GitHub. manejo de roles en los usuarios

#### Inty360:
Parece también apta para Geomatica, respecto a la creación de formularios y el servicio offline, permite el ingreso de información como fotos o videos, presenta también datos geográficos de los datos recolectados, se pueden exportar los datos a distintos softwares y formatos, gestión de usuarios, funciona tanto en Android como en iOS, y en un pc desde cualquier navegador, tiene una versión de escritorio para la visualización de los datos. La creación de los formularios y la asignación de los roles se realiza desde la página web. los planes de pago son mensuales y varían respecto al tamaño de almacenamiento en la nube, y otras limitaciones respecto a la cantidad de proyectos y preguntas. no tiene GitHub.

#### Moreapp:
Permite la creación de formularios, trabajo offline, personalización de las encuestas, gestión de usuarios, exportar e importar, cuenta con una API y permite la integración con otros softwares, tiene una versión gratuita con las cosas básicas, las de pago se manejan comprando "créditos" los cuales se van gastando al uso de las funciones, rellenado de formularios complejos, etc. la parte de la creación de formularios y demás se realiza en la web.

#### Twine:
No presenta mucha información en su página de inicio, da indicios de poder descargar la herramienta, pero para hacerlo, y para acceder a cualquier sección de la página, hay que iniciar sesión.

#### Formhub:
Como se mencionó anteriormente la "versión" instalable que se menciona en ONA se trata de formhub y pues esta aplicación cumple los requisitos mencionados, sin embargo, también función usando la plataforma, no parecen existir planes y como se dijo en la página no parece haber indicios de una instalación resiente de formhub.

#### Smap:
la versión de servidor parece requerir de instalaciones extras por ejemplo un editor de formularios como XLSForm, aunque existe un editor online que facilita el proceso, al parecer es una posibilidad para Geomatica, el pago es más que todo por el soporte o el alquiler de un servidor. presenta una buena documentación, pero el uso parece ser algo complicado. habría que revisar. tiene GitHub, con un gran porcentaje en JS, no explica como instalar, pero estas se encuentran en los archivos de documentación.

#### Mfieldwork:
Parece manejar un sistema de créditos y se comprarían dichos créditos, es una plataforma, no cuenta con alguna versión de escritorio y los datos quedan guardados en los servidores de dicha empresa.

#### Datawinners: 
Cuenta con pagos mensuales, no tiene versión de escritorio, es web. no tiene GitHub.

#### ODK: 
Más que una plataforma o un software, es un conjunto de herramientas que ayudan en las distintas tareas de recolección de datos, si se decide por el uso de las mismas habría que configurarlas e integrarlas entre sí. muchas de las paginas anteriores usan algunos de los softwares de ODK sobre todo para la recolección de datos en móviles. presenta una guía de uso dando explicación que se requieren 3 pasos, diseñar, colectar y agregar, el diseño se puede realizar utilizando alguna herramienta xlsForm o la herramienta presentada en la página, la recolección se realiza en un dispositivo Android y la configuración con el servidor va con una de las herramientas de ODK. parece que se requiere el uso de las credenciales de google.

#### Kobotoolbox: 
Python, permite instalación local, permite la recolección de datos con mapas, tiene GitHub, función offline, es código abierto, pero no se presenta con un único repositorio, si se quiere instalar en una maquina local requiere la instalación de varios de los repositorios presentes en el GitHub de la aplicación. los desarrolladores de kobotoolbox recomiendan el uso de dockers que la instalación manual.
